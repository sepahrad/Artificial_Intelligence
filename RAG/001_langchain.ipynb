```python

OPENAI_API_KEY = ""
MODEL = "gpt-5"
```

```python
from langchain_perplexity import ChatPerplexity
from langchain_community.document_loaders import PyPDFLoader
```

```python
model = ChatPerplexity(
    temperature=0, 
    pplx_api_key=OPENAI_API_KEY, 
    model="sonar"
)
response = model.invoke("Hello! How are you?")
print(response)
#print(response['text'])
#print(response.content)
```

```python
file_path = "./Stock-Investing-101-eBook.pdf"
loader = PyPDFLoader(file_path)
documents = loader.load()
print(f"Number of pages: {len(documents)}")
print(documents[0].page_content)
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
parsed_output = output_parser.parse("This is a sample output.")
```

```python
chain = model | output_parser
print(chain.invoke("Hello! How are you?"))
```

```python
from langchain_core.prompts import PromptTemplate
template = """
Answer the question based on the context below.If you cannot answer the question using the context, please respond with "I don't know".

Context: {context}
Question: {question}

"""
```

```python
prompt = PromptTemplate.from_template(template)
prompt.format(context="As a financial advisor, asnwer concisely.",
              question="How to invest in stocks?")

chain = prompt | model | output_parser
print(chain.invoke({"context": "As a financial advisor, asnwer concisely.","question": "How to invest in stocks?"}))
```
