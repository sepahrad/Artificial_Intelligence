```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
x
```

```python
y = (x ** 2 - 1 ) / (x - 1)
y
```

```python
plt.plot(x, y)
plt.axvline(1, color='r', linestyle='--')
plt.axhline(2, color='r', linestyle='--')
plt.show()
```
