```python
import numpy as np
```

```python
def gradient_descent(gradient, start, learn_rate, n_iter, tolerance=1e-06):
  for i in range(n_iter):
    s = -learn_rate * gradient(start)
    if np.all(np.abs(s) <= tolerance):
      break
    start += s
  return start
```

```python
g = lambda x: 2 * x
```

```python
r = gradient_descent(g, 10, 0.2, 25)
r
```

```python
np.round(r)
```

```python
r = gradient_descent(g, 10, 0.8, 25)
r
```

```python
np.round(r)
```

```python
r = gradient_descent(g, 10, 0.009, 1000)
r
```

```python
np.round(r)
```

```python
g = lambda x : np.array([2 * x[0],
                         4 * x[1] ** 3])
```

```python
r = gradient_descent(g, start=[1,1], learn_rate=0.2, n_iter=50)
r
```

```python
np.round(r)
```
