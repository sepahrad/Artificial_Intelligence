```python
import sympy as sp
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
f = x ** 3 + 2 * x + 3
```

```python
f
```

```python
sp.diff(f, x)
```

```python
sp.diff(f)
```

```python
f = x ** 3 + 2 * x + 3 + 2 * y**2
```

```python
sp.diff(f, y)
```

```python
sp.diff(f, x)
```

```python
f = sp.sin(x**3 + 2 * x)
```

```python
sp.diff(f)
```

```python
f = sp.log(x**3 + 2 *y)
```

```python
sp.diff(f, x)
```

```python
sp.diff(f, y)
```

```python
f = sp.exp(3 ** x)
```

```python
sp.diff(f, x)
```

```python
from sympy import Function
```

```python
f = Function('f')
u = f(x, y)
u
```

```python
ux = sp.diff(u, x)
ux
```

```python
uy = sp.diff(u, y)
uy
```

```python
from sympy import Eq
Eq(1 - (ux/x) + (uy/u), 5)
```
