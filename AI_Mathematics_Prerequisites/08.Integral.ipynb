```python
import sympy as sp
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
sp.integrate(x)
```

```python
sp.integrate(x, x)
```

```python
sp.integrate(x, y) #Integ(xdy)
```

```python
sp.integrate(x**2, x)
```

```python
sp.integrate(x**2, y)
```

```python
f = x ** 2 + 3 * y
f
```

```python
sp.integrate(f, x)
```

```python
sp.integrate(f, y)
```

```python
sp.integrate(sp.exp(-2 * x), x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```
