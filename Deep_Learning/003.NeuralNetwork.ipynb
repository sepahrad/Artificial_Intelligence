```python
import tensorflow as tf

# Data
x = tf.constant([0.04, 0.2, 0.34, 0.46, 0.51, 0.66, 0.78, 1.02, 1.3, 1.42, 1.65, 1.76, 1.88, 1.98], dtype=tf.float32)
y = tf.constant([0.67, 0.87, 1.08, 1.1, 1.5, 1.3, 1.54, 1.55, 1.78, 1.98, 1.86, 1.6, 1.9, 1.99], dtype=tf.float32)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation='linear')
])

# Compile
model.compile(loss=tf.keras.losses.MeanSquaredError(), optimizer=tf.keras.optimizers.SGD())

# Train
model.fit(x, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Predict
x_values = np.linspace(0, 2, 100)
y_pred = model.predict(x_values)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Actual')
plt.plot(x_values, y_pred, color='red', label='Predicted')
plt.xlim([0, 2])
plt.ylim([0, 2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```
